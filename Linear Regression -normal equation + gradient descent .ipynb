## feature scaling for gradient descent 

```python

X=np.array([[1,89,7921],[1,72,5184],[1,94,8836],[1,69,4761]]);
n,m=X.shape
moy=[np.sum(X[:,i])/n for i in range(m)]
#max=np.max(X[:,1])
#print(max)
moy[0]=0
print(moy)
var=[ np.max(X[:,i])-np.min(X[:,i]) for i in range(m)]
var[0]=1
print(var)
Y=np.array([96,74,87,78])
X=[[(X[i,j]-moy[j])/var[j] for j in range (m)]for i in range (n)]
X=np.array(X)
print(X)
 
```

```python
plt.plot(X[:,1], Y,linestyle = 'none',  marker = 'o')

plt.show() # affiche la figure a l'en
```

# equation de normalisation


```python
def theta(X,Y):
    inv=np.dot(X.T,X)
    inv=np.linalg.inv(inv)
    theta=(np.dot(inv,X.T)).dot(Y)
    return theta

theta=theta(X,Y)
print(theta)
```

```python
A=theta.dot(np.transpose(X[1,:]))
print(A)
```

```python
l= [-1,-0.5,0,0.5,1]
val=[np.dot(theta,(np.array([1,x,x*x])).T) for x in l]
plt.plot(val,np.array(l),linestyle = 'none',  marker = 'o')
plt.plot(X[:,1], Y,linestyle = 'none',  marker = 'o',color = 'green')

plt.show()
```

```python
thet=[random() for i in range(3)]
print(thet)
thet=np.array([thet])
print(type(thet))
```

```python
alpha=2
print(thet[0,0])
def newtheta(theta,X,Y):
    J=0
    t,l=theta.shape
    for j in range(n):
        J+=(np.dot(theta,X[j,:])-Y[j])
    J=J/(n)
    theta[t-1,0]=theta[t-1,0] -alpha*J
    for i in range(l):
        J=0
        for j in range(n):
            J+=(np.dot(theta,X[j,:])-Y[j])*X[j,i]
        J=J/(n)
        theta[t-1,i]=theta[t-1,i]-alpha*J
    return theta


theti=newtheta(thet,X,Y)
print(theti)
     
```
